# Loc vs iloc
V dnesnej lekcii si prejdeme moznosti filtrovania dataframu na zaklade funkcii loc a iloc.
- __loc__ sa pouziva na filtrovanie stlpcov alebo riadkov na zaklade presnej hodnoty
- __iloc__ sa pouziva na filtrovanie stlpcov alebo riadkov na zaklade indexu (indexujeme od 0). Pristup je rovnaky ako v pripade poli

Funkcie loc a iloc su ovela efektivnejsie a lepsie ako pristup pomocou hranatych zatvoriek!!!!

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(1, index=[1, 3, 2, 4], columns=["a", "b", 1, 2])
df = pd.DataFrame(1, index=["i", "j", "a", "b"], columns=["a", "b", 1, 2])
```

```python
df1
```

```python
# riadok s indexom 2
df1.loc[2, :] # vystupom je riadok ako pandas series (to ze je to v stlpci vas moze mylit)
```

```python
# riadky s indexami 1 a 2 a stlpec a
df1.loc[[1,2], "a"]
```

```python
# POZOR toto moze byt velmi matuce :2 znamena ze chcem riadky od zaciatku az po hodnotu indexu 2 (je tam aj 3 lebo bola pred 2)
# chcem stlpce ["a", "b", 2]
df1.loc[:2, ["a", "b", 2]]
```

```python
# Pomocou locu vieme spravit aj napriklad toto !! TOTO JE MATUCE OPROTI PRISTUPU V POLIACH
df1.loc[:2, :"b"]
```

```python
df1.iloc[-2:-1, 1:]
```
