# Multyindex
Okrem jednoduchych indexov a jednoduchych stlpcov python umoznuje multyindex v pripade indexu aj stlpcu.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(
    data={
        ("player1", "x"): np.arange(2,7),
        ("player1", "y"): np.arange(2,7),
        ("player2", "x"): np.arange(2,7),
        ("player2", "y"):np.arange(2,7),
    },
    index=[("player1", "x"),("player1", "x"),("player1", "x"),("player1", "x"),("player1", "x")]
)
df
```

```python
df["player1"]["x"]
```

```python
df.loc[:, ("player1", "x")]
```

```python
df.loc[:, ("player1", "x")] = 1
df
```

# Uprava dataframu
Je beznou praxou pocas prace s dataframom pridavat nove riadky alebo stlpce. V dnesnej lekcii si ukazeme ako na to.

```python
df = pd.DataFrame(data = {"a": np.arange(1,10), "b": np.arange(10,19)})
df
```

```python
df["c"]=np.arange(20,29)
df
```

```python
#(a+b)**2
df["d"]=(df["a"] + df["b"])**2
df
```

```python
df.loc[9] = [0,1,2,3]
df
```

```python
df.loc["misko"] = [0,1,2,3]
df
```
